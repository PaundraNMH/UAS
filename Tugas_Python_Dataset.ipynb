```python
#@title
# Posisi Plotting
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))
 
# Persiapan Database ( Rumus Khusus untuk Penyesuaian Grafik )
AGE = [35, 37, 51, 57, 67] # Nilai Rata Rata Setiap Value yang Muncul
HEART_DISEASE = [30, 40, 50, 60, 70] # Nilai dikurang 20 mengikuti total data yang terfilterisasi di masing masing nilai dengan total data yaitu 2050 data
SMOKING_STATUS = [30, 40, 50, 60, 70] # Sama Seperti Atribut Heart Disease
 
# Atur Posisi Bar
br1 = np.arange(len(AGE))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
 
plt.bar(br1, AGE, color ='r', width = barWidth,
        edgecolor ='grey', label ='AGE')
plt.bar(br2, HEART_DISEASE, color ='g', width = barWidth,
        edgecolor ='grey', label ='HEART_DISEASE')
plt.bar(br3, SMOKING_STATUS, color ='b', width = barWidth,
        edgecolor ='grey', label ='SMOKING_STATUS')
 
plt.xlabel('LEVEL STATUS = Level 1-5', fontweight ='bold', fontsize = 15)
plt.ylabel('HEART_DISEASE = ', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(AGE))],
        ['170', '177', '190', '200', '220'])
 
plt.legend()
plt.show()
```

```python
data = {'age':60, 'heart_disease':70, 'level_status':40,
        'smoking_status':50} # Nilai Atribut Disesuaikan dengan format grafik dengan nilai lebih dari satuan
courses = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (10, 5))
 
plt.bar(courses, values, color ='skyblue',
        width = 0.4)

plt.xlabel("Variable Rate")
plt.ylabel("Range Based on Variable")
plt.title("Result on Stroke Possibility Test")
plt.show()
```
